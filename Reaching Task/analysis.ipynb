```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
error_angles = pd.read_csv('error_angles_exp2.csv', header=None)
error_angles = error_angles.to_numpy(dtype=float).flatten()
len(error_angles)

move_faster_events = pd.read_csv('move_faster_exp2.csv', header=None)
move_faster_events = move_faster_events.to_numpy(dtype=bool).flatten()
len(move_faster_events)

attempts = np.arange(1, len(error_angles)+1)
```

```python
fig = plt.figure(figsize=(15, 5))

plt.plot(attempts[~move_faster_events], error_angles[~move_faster_events], 'o', linestyle="--", color = "blue", label = "Fast Enough")
plt.plot(attempts[move_faster_events], error_angles[move_faster_events], 'x', linestyle="", color = "red", label = "Too Slow")

plt.text(1, plt.ylim()[1] + 1, 'right hand\nno pertubation', color='black', verticalalignment='top', horizontalalignment='left', weight='bold')
plt.axvline(x=40, color='black', linestyle='-')
plt.text(41, plt.ylim()[1] + 1, 'CONTROL:\nright hand\ngradual pertubation\n+30°', color='black', verticalalignment='top', horizontalalignment='left', weight='bold')
plt.axvline(x=80, color='black', linestyle='-')
plt.text(81, plt.ylim()[1] + 1, 'right hand\ngradual pertubation\n', color='black', verticalalignment='top', horizontalalignment='left', weight='bold')
plt.axvline(x=120, color='black', linestyle='-')
plt.text(121, plt.ylim()[1] + 1, 'UNLEARNING:\nright hand\ngradual pertubation\n-30°', color='black', verticalalignment='top', horizontalalignment='left', weight='bold')
plt.axvline(x=160, color='black', linestyle='-')
plt.text(161, plt.ylim()[1] + 1, 'left hand\nno pertubation', color='black', verticalalignment='top', horizontalalignment='left', weight='bold')
plt.axvline(x=200, color='black', linestyle='-')
plt.text(201, plt.ylim()[1] + 1, 'ADAPTATION:\nleft hand\ngradual pertubation\n+30°', color='black', verticalalignment='top', horizontalalignment='left', weight='bold')
plt.axvline(x=240, color='black', linestyle='-')
plt.text(241, plt.ylim()[1] + 1, 'right hand\nno pertubation', color='black', verticalalignment='top', horizontalalignment='left', weight='bold')
plt.axvline(x=280, color='black', linestyle='-')
plt.text(281, plt.ylim()[1] + 1, "TRANSFER?:\nright hand\ngradual pertubation\n+30°", color='black', verticalalignment='top', horizontalalignment='left', weight='bold')

plt.ylim(plt.ylim()[0], plt.ylim()[1] + 1.5)
plt.xlim(0, 320)
plt.grid()
plt.xlabel('Attempt')
plt.ylabel('Error Angle [°]')
plt.legend(title='Movement Speed', loc = "lower right")
plt.tight_layout()
```
